```python
import geopandas as gpd
import sys
sys.path.append('..')
from utils.deafrica_utils.deafrica_spatialtools import xr_rasterize
```

```python
land_class_data = gpd.read_file('landuse_DK.shp')[['class', 'geometry']]
```

**Change the CRS to EPSG:4326**

```python
land_class_data = land_class_data.to_crs("EPSG:4326")
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 8))
land_class_data.plot(column='class', legend=True, ax=ax)
plt.title('Land Classes')
plt.show()
```
